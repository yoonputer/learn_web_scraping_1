```python
from selenium import webdriver
```

```python
browser = webdriver.Chrome('./chromedriver.exe')
```

```python
browser.get('http://ticket.interpark.com/Contents/Ranking?smid1=s_menu&smid2=performance')
```

```python
html = browser.page_source
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html,'html.parser')
```

```python
len(soup),type(soup)
```

```python
tags = soup.select('div.items > ul > li')
```

```python
len(tags), type(tags)
```

```python
tag = tags[0]
```

```python
title = tag.select('a.prdName')
len(title), type(title)
```

```python
for tag in tags:
    title = tag.select('a.prdName')
    data_range = tag.select('a.prdDuration')
    print(title[0].text.strip(),data_range[0].text.strip())
```

```python
contents = list()
for tag in tags:
    title = tag.select('a.prdName')
    data_range = tag.select('a.prdDuration')
    interpark = [title[0].text.strip(),data_range[0].text.strip()]
    contents.append(interpark)
    
contents
```

```python
import pandas as pd
```

```python
pd_data = pd.DataFrame(contents, columns = ['title','Data_Range'])
```

```python
pd_data.to_excel('./saves/interpark_scraping.xls',index = False)
```
